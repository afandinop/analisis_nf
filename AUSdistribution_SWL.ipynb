#### Distribución estaciones válidas Australia

```python
fig = plt.figure(figsize=(15, 12))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

ax.set_extent([112, 154, -44, -10], crs=ccrs.PlateCarree())

ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle='-')

ax.scatter(lon_wa, lat_wa, transform=ccrs.PlateCarree(), 
           color='blue', s=10, alpha=0.7, label='Estaciones WA')

ax.scatter(lon_nsw, lon_nsw, transform=ccrs.PlateCarree(), 
           color='red', s=10, alpha=0.7, label='Estaciones NSW')

ax.scatter(lon_nt, lat_nt, transform=ccrs.PlateCarree(), 
           color='purple', s=10, alpha=0.7, label='Estaciones NT')

ax.scatter(lon_qld, lat_qld, transform=ccrs.PlateCarree(), 
           color='yellow', s=10, alpha=0.7, label='Estaciones QLD')

ax.scatter(lon_tas, lat_tas, transform=ccrs.PlateCarree(), 
           color='pink', s=10, alpha=0.7, label='Estaciones TAS')

ax.scatter(lon_vic, lat_vic, transform=ccrs.PlateCarree(), 
           color='green', s=10, alpha=0.7, label='Estaciones VIC')

ax.scatter(lon_sa, lat_sa, transform=ccrs.PlateCarree(), 
           color='brown', s=10, alpha=0.7, label='Estaciones SA')

ax.set_title('Distribución de Estaciones en Australia (SWL)')
ax.legend()

# Añadir líneas de cuadrícula
gl = ax.gridlines(draw_labels=True, linestyle='--')
gl.top_labels = False
gl.right_labels = False

# Contar el número total de estaciones
total_estaciones = len(lat_wa) + len(lat_nsw) + len(lat_nt) + len(lat_qld) + len(lat_tas) + len(lat_vic) + len(lat_sa)
ax.text(0.05, 0.95, f'Número total de estaciones: {total_estaciones}', 
        transform=ax.transAxes, fontsize=10, 
        verticalalignment='top', bbox=dict(facecolor='white', alpha=0.7))

plt.show()
```
